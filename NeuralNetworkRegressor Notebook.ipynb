```python
import random,math
data = []
for i in range(1000):
        try:
            #x = random.randint(150,350)
            #y = random.randint(100,200)
            #w= random.randint(200,400)
            x = .01*i + 190
            y = 120 + random.randint(-10,10)
            w = random.randint(50,200)

            dire = math.atan((float(w)/15)/(-.2*float(x-240)))
            
            #dire=math.atan(float(y+120)/(x-160))
            data.append({"X":x,"Y":y,"Width":w,"Direction":dire})
        except:
                continue
```

```python
import pandas as pd
from sklearn.neural_network import MLPRegressor

df = pd.DataFrame(data)
X = df.iloc[:,:3].values
Y = df["Direction"].values
clf = MLPRegressor(hidden_layer_sizes=(10,10,10),max_iter = 50000)
print (clf.fit(X,Y))
```

```python
print ("Coefficients: ", clf.coefs_ )
print("Predicted Value: ", clf.predict([[150,190,90]]))
```
